```python


# Importar librerÃ­as
import whisper
import os

# Cargar el modelo de Whisper (elige "base" o "small" segÃºn el rendimiento que necesites)
model = whisper.load_model("base")

# Transcribir el archivo MP4
result = model.transcribe("audio.mp4")

# Ruta del archivo transcription.txt en el entorno de Colab
transcription_path = "/content/transcription.txt"

# Guardar la transcripciÃ³n en el archivo de texto
with open(transcription_path, "w", encoding="utf-8") as file:
    file.write(result["text"])

# Verificar que el archivo existe
if os.path.exists(transcription_path):
    print(f"âœ… TranscripciÃ³n guardada en {transcription_path}")
else:
    print("âŒ Error: No se pudo guardar el archivo")

# Mostrar la transcripciÃ³n en pantalla
#print("\nğŸ“„ Contenido de la transcripciÃ³n:\n", result["text"])
```
