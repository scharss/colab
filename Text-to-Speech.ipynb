```python
import edge_tts
import nest_asyncio
import asyncio

# Aplicar el parche para usar asyncio en Colab
nest_asyncio.apply()

async def text_to_mp3(input_file, output_file, voice):
    try:
        # Leer el contenido del archivo
        with open(input_file, 'r', encoding='utf-8') as file:
            text = file.read()
            if text.strip():  # Verificar si el archivo tiene contenido
                print(f"Generando archivo de audio '{output_file}'...")

                # Inicializar Edge TTS
                communicate = edge_tts.Communicate(text, voice)
                await communicate.save(output_file)

                print(f"El archivo de audio se ha guardado como '{output_file}'.")
            else:
                print("El archivo está vacío.")
    except FileNotFoundError:
        print(f"No se encontró el archivo '{input_file}'. Asegúrate de que esté en el mismo directorio.")
    except Exception as e:
        print(f"Ocurrió un error: {e}")

# Ejecutar la función asincrónica
input_file = "article.txt"
output_file = "article_audio.mp3"
voice = "es-MX-JorgeNeural"  # Voz masculina latinoamericana

# Llamada segura en Colab
await text_to_mp3(input_file, output_file, voice)
```
